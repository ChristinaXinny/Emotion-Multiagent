## Test Perception Agent (FinBERT)

```python
# Initialize Perception Agent
perception_config = {
    "model_name": "ProsusAI/finbert",
    "device": "cpu"
}

perception_agent = PerceptionAgent(config=perception_config)
print("Perception Agent initialized")
```

```python
# Test with sample texts
test_texts = [
    "Apple reports record quarterly earnings, beating analyst expectations.",
    "Stock prices plummet amid concerns about economic recession.",
    "Company announces merger with competitor in strategic move."
]

for text in test_texts:
    result = perception_agent.process(text)
    print(f"\nText: {text}")
    print(f"Sentiment: {result['sentiment']}")
    print(f"Confidence: {result['confidence']:.3f}")
    print(f"Scores: {result['scores']}")
```
